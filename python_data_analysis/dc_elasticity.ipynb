```python
import matplotlib.pyplot as plt
import numpy as np

# x값을 지수적으로 증가 (2^n)
n = np.arange(1, 11, 1)  # 1부터 10까지
x_exp = 2 ** n
y_log = np.log(x_exp)
slope_exp = 1 / x_exp

# 로그 함수 그래프 (x가 지수적으로 증가)
plt.figure(figsize=(10, 5))
plt.plot(x_exp, y_log, marker='o', label="y = log(x), x=2^n", linewidth=2)
# plt.title("x가 지수적으로 증가할 때 로그곡선")
plt.xlabel("x (2^n)")
plt.ylabel("log(x)")
plt.grid(True)
plt.legend()
plt.show()

# 기울기 그래프 (1/x, x가 지수적으로 증가)
plt.figure(figsize=(10, 5))
plt.plot(x_exp, slope_exp, marker='o', label="기울기 (dy/dx = 1/x), x=2^n", linewidth=2)
# plt.title("x가 지수적으로 증가할 때 로그곡선의 기울기 (1/x)")
plt.xlabel("x (2^n)")
plt.ylabel("slope (1/x)")
plt.grid(True)
plt.legend()
plt.show()
```
