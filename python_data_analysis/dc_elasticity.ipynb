```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# 한글 폰트 설정 (운영체제에 맞게 선택)
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우
# plt.rcParams['font.family'] = 'AppleGothic'   # 맥
# plt.rcParams['font.family'] = 'NanumGothic'   # 리눅스

# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

n = np.arange(1, 11, 1)  # 1부터 10까지
x_exp = 2 ** n
y_log = np.log(x_exp)
slope_exp = 1 / x_exp

plt.figure(figsize=(10, 5))
plt.plot(y_log, x_exp, marker='o', label="y = 2^n, log(y) (x축)", linewidth=2)
plt.title("로그 값의 선형 증가에 따른 지수 함수의 증가")
plt.xlabel("x축 = log(y) (로그 값)")
plt.ylabel("y축 = 2^n (지수 값)")
plt.grid(True)
plt.legend()
plt.annotate('로그 값이 조금만 증가해도', xy=(2, 8), xytext=(1, 300),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10)
plt.annotate('y는 지수적으로 크게 증가', xy=(6, 450), xytext=(4.3, 750),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10)
plt.show()
```
