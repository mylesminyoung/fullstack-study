```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

#    (  )
plt.rcParams['font.family'] = 'Malgun Gothic'   # 
# plt.rcParams['font.family'] = 'AppleGothic'   # 
# plt.rcParams['font.family'] = 'NanumGothic'   # 

#    
plt.rcParams['axes.unicode_minus'] = False

n = np.arange(1, 11, 1)  # 1 10
x_exp = 2 ** n
y_log = np.log(x_exp)
slope_exp = 1 / x_exp

plt.figure(figsize=(10, 5))
plt.plot(y_log, x_exp, marker='o', label="y = 2^n, log(y) (x)", linewidth=2)
plt.title("       ")
plt.xlabel("x = log(y) ( )")
plt.ylabel("y = 2^n ( )")
plt.grid(True)
plt.legend()
plt.annotate('   ', xy=(2, 8), xytext=(1, 300),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10)
plt.annotate('y   ', xy=(6, 450), xytext=(4.3, 750),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10)
plt.show()
```
