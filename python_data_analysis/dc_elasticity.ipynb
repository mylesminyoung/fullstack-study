```python
import matplotlib.pyplot as plt
import numpy as np

# x   (2^n)
n = np.arange(1, 11, 1)  # 1 10
x_exp = 2 ** n
y_log = np.log(x_exp)
slope_exp = 1 / x_exp

#    (x  )
plt.figure(figsize=(10, 5))
plt.plot(x_exp, y_log, marker='o', label="y = log(x), x=2^n", linewidth=2)
# plt.title("x    ")
plt.xlabel("x (2^n)")
plt.ylabel("log(x)")
plt.grid(True)
plt.legend()
plt.show()

#   (1/x, x  )
plt.figure(figsize=(10, 5))
plt.plot(x_exp, slope_exp, marker='o', label=" (dy/dx = 1/x), x=2^n", linewidth=2)
# plt.title("x      (1/x)")
plt.xlabel("x (2^n)")
plt.ylabel("slope (1/x)")
plt.grid(True)
plt.legend()
plt.show()
```
